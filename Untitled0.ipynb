```python
use_columns = ['Release Period', 'Whether Remake', 'Whether Franchise', 'Genre', 'New Actor', 'New Director', 'New Music Director']
y = train['Revenue(INR)']
train = train[use_columns]
test = test[use_columns]
```

```python
train.head()
```

```python
#カテゴリ変数のダミー変数化
train = pd.get_dummies(train, drop_first=True)
test = pd.get_dummies(test, drop_first=True)
```

```python
display(train.head(), test.head())
```

```python
print(train.dtypes)
print(test.dtypes)
```

```python
train = pd.get_dummies(train, drop_first=True).astype(int)
test = pd.get_dummies(test, drop_first=True).astype(int)
```

```python
display(train.head(), test.head())
```

```python
print(train.shape)
print(test.shape)
```

```python
train = train.drop('Genre_documentary', axis=1)
print(train.shape)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train, y, test_size=0.25, random_state=82)
```

```python
ls = Lasso(alpha=0.5)
ls.fit(X_train, y_train)
```

```python
#学習
ls.fit(X_train, y_train)

#予測・精度算出
train_pred = ls.predict(X_train)
valid_pred = ls.predict(X_valid)
print('train score : ', mean_absolute_error(y_train, train_pred))
print('valid score : ', mean_absolute_error(y_valid, valid_pred))
```

```python
fig, ax = plt.subplots()
ax.axline((0,0), slope=1.0, color='red') #実測値と予測値が一致している場合のライン
plt.scatter(y_train,train_pred)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.axline((0,0), slope=1.0, color='red') #実測値と予測値が一致している場合のライン
plt.scatter(y_valid,valid_pred)
plt.show()
```

```python
# testデータの予測
predict = ls.predict(test)
```
